# 圧力データ比較

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
#data_p = np.loadtxt(f'./expdata/9449af_05f4aa0eaa1d4164a1024d07c562592a/SPHERIC_TestCase10/data_files/lateral_water_1x.txt')
df_p1_exp = pd.read_table(f"../expdata/9449af_05f4aa0eaa1d4164a1024d07c562592a/SPHERIC_TestCase10/data_files/lateral_water_1x.txt").iloc[:,0:2]#, columns=['time','p1','p2','p3','p4','p5','p6'])  
df_p1_exp['Pressure[Pa]'] = df_p1_exp['Pressure[mbar]']*100

caedata_p = np.loadtxt(f'./postProcessing/probes/0/p')
df_p1_cae = pd.DataFrame(caedata_p, columns=['time','CoGp', 'p1','p2','p3','p4','p5','p6']) 
df_p1_cae['p1(Pa)abs'] = df_p1_cae['p1']-df_p1_cae['p1'][0]

plt.title('p1 pressure',fontsize=24)

plt.plot(df_p1_exp.iloc[:,0],df_p1_exp.iloc[:,2],linestyle='solid',color='black',label="Exp")
plt.plot(df_p1_cae.iloc[:,0],df_p1_cae.iloc[:,8],linestyle='solid',color='red',label="OpenFOAM")
plt.xlabel('time(sec)',fontsize=16)
plt.ylabel('Pressure(Pa)',fontsize=16)
plt.legend(fontsize=24)
plt.grid()
plt.xticks(list(np.arange(0,8,1)),fontsize=12)
plt.yticks(list(np.arange(-1000,9000,1000)),fontsize=12)
```
